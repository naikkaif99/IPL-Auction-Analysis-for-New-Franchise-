```python
matches["season"] = pd.to_numeric(matches["season"], errors="coerce")
matches = matches.dropna(subset=["season"])
matches["season"] = matches["season"].astype(int)

matches_2020_2025 = matches[
    (matches["season"] >= 2020) & (matches["season"] <= 2025)
].copy()

match_ids = matches_2020_2025["id"].unique()

deliveries_2020_2025 = deliveries[
    deliveries["match_id"].isin(match_ids)
].copy()

# =========================
# 2. BASIC COLUMN HANDLING
# =========================
if "batter" not in deliveries_2020_2025.columns and "batsman" in deliveries_2020_2025.columns:
    deliveries_2020_2025["batter"] = deliveries_2020_2025["batsman"]

if "total_runs" not in deliveries_2020_2025.columns:
    deliveries_2020_2025["total_runs"] = (
        deliveries_2020_2025.get("batsman_runs", 0) +
        deliveries_2020_2025.get("extra_runs", 0)
    )

if "player_dismissed" not in deliveries_2020_2025.columns:
    deliveries_2020_2025["player_dismissed"] = np.nan

if "dismissal_kind" not in deliveries_2020_2025.columns:
    deliveries_2020_2025["dismissal_kind"] = np.nan

# =========================
# 3. BATSMAN ANALYSIS
# =========================
if "extras_type" in deliveries_2020_2025.columns:
    batter_balls_df = deliveries_2020_2025[
        deliveries_2020_2025["extras_type"] != "wides"
    ].copy()
else:
    batter_balls_df = deliveries_2020_2025.copy()

batting_stats = batter_balls_df.groupby("batter").agg(
    runs=("batsman_runs", "sum"),
    balls=("batter", "count"),
    innings=("match_id", pd.Series.nunique),
    fours=("batsman_runs", lambda x: (x == 4).sum()),
    sixes=("batsman_runs", lambda x: (x == 6).sum())
).reset_index()

dismissal_exclude = ["run out", "retired hurt", "obstructing the field"]

dismissals_df = deliveries_2020_2025[
    deliveries_2020_2025["player_dismissed"].notna()
].copy()

dismissals_df = dismissals_df[
    ~dismissals_df["dismissal_kind"].isin(dismissal_exclude)
]

dismissals_count = dismissals_df.groupby("player_dismissed").size().reset_index(name="dismissals")
dismissals_count.rename(columns={"player_dismissed": "batter"}, inplace=True)

batting_stats = batting_stats.merge(dismissals_count, on="batter", how="left")
batting_stats["dismissals"] = batting_stats["dismissals"].fillna(0)

batting_stats["strike_rate"] = np.where(
    batting_stats["balls"] > 0,
    (batting_stats["runs"] / batting_stats["balls"]) * 100,
    0
)

batting_stats["average"] = np.where(
    batting_stats["dismissals"] > 0,
    batting_stats["runs"] / batting_stats["dismissals"],
    batting_stats["runs"]
)

season_map = matches_2020_2025[["id", "season"]].copy()
season_map.rename(columns={"id": "match_id"}, inplace=True)

bat_season = deliveries_2020_2025.merge(season_map, on="match_id", how="left")
bat_season = bat_season.groupby(["batter", "season"]).agg(
    season_runs=("batsman_runs", "sum")
).reset_index()

consistency_stats = bat_season.groupby("batter").agg(
    seasons_played=("season", "nunique"),
    avg_season_runs=("season_runs", "mean")
).reset_index()

batting_stats = batting_stats.merge(consistency_stats, on="batter", how="left")
batting_stats = batting_stats.fillna(0)

batting_stats = batting_stats[batting_stats["runs"] >= 150].copy()

batting_stats["batting_score"] = (
    batting_stats["runs"] * 0.40 +
    batting_stats["strike_rate"] * 0.25 +
    batting_stats["average"] * 0.20 +
    batting_stats["avg_season_runs"] * 0.15
)

# =========================
# 4. BATTING POSITION
# =========================
innings_batters = deliveries_2020_2025.groupby(
    ["match_id", "inning", "batting_team", "batter"]
).agg(
    first_over=("over", "min"),
    first_ball=("ball", "min")
).reset_index()

innings_batters = innings_batters.sort_values(
    by=["match_id", "inning", "first_over", "first_ball"]
)

innings_batters["batting_position"] = innings_batters.groupby(
    ["match_id", "inning", "batting_team"]
).cumcount() + 1

bat_pos_stats = innings_batters.groupby("batter").agg(
    avg_batting_position=("batting_position", "mean"),
    opening_innings=("batting_position", lambda x: ((x == 1) | (x == 2)).sum()),
    total_innings=("batting_position", "count")
).reset_index()

bat_pos_stats["opening_ratio"] = (
    bat_pos_stats["opening_innings"] / bat_pos_stats["total_innings"]
)

batting_stats = batting_stats.merge(bat_pos_stats, on="batter", how="left")
batting_stats = batting_stats.fillna(0)

# =========================
# 5. BOWLING ANALYSIS
# =========================
if "extras_type" in deliveries_2020_2025.columns:
    legal_balls_df = deliveries_2020_2025[
        deliveries_2020_2025["extras_type"] != "wides"
    ].copy()
else:
    legal_balls_df = deliveries_2020_2025.copy()

non_bowler_dismissals = ["run out", "retired hurt", "obstructing the field"]

wickets_df = deliveries_2020_2025[
    deliveries_2020_2025["player_dismissed"].notna()
].copy()

wickets_df = wickets_df[
    ~wickets_df["dismissal_kind"].isin(non_bowler_dismissals)
]

bowling_stats = legal_balls_df.groupby("bowler").agg(
    balls_bowled=("bowler", "count"),
    runs_conceded=("total_runs", "sum"),
    dot_balls=("total_runs", lambda x: (x == 0).sum())
).reset_index()

wicket_counts = wickets_df.groupby("bowler").size().reset_index(name="wickets")
bowling_stats = bowling_stats.merge(wicket_counts, on="bowler", how="left")
bowling_stats["wickets"] = bowling_stats["wickets"].fillna(0)

bowling_stats["overs"] = bowling_stats["balls_bowled"] / 6
bowling_stats["economy"] = np.where(
    bowling_stats["overs"] > 0,
    bowling_stats["runs_conceded"] / bowling_stats["overs"],
    0
)
bowling_stats["dot_ball_pct"] = np.where(
    bowling_stats["balls_bowled"] > 0,
    (bowling_stats["dot_balls"] / bowling_stats["balls_bowled"]) * 100,
    0
)

bowling_stats = bowling_stats[bowling_stats["balls_bowled"] >= 120].copy()

bowling_stats["bowling_score"] = (
    bowling_stats["wickets"] * 0.45 +
    (100 / (bowling_stats["economy"] + 1)) * 0.25 +
    bowling_stats["dot_ball_pct"] * 0.20 +
    bowling_stats["overs"] * 0.10
)

# =========================
# 6. ALL-ROUNDER ANALYSIS
# =========================
all_rounders = batting_stats.merge(
    bowling_stats,
    left_on="batter",
    right_on="bowler",
    how="inner"
).copy()

all_rounders["player"] = all_rounders["batter"]

all_rounders = all_rounders[
    (all_rounders["runs"] >= 150) &
    (all_rounders["wickets"] >= 8)
].copy()

all_rounders["allrounder_score"] = (
    all_rounders["batting_score"] * 0.5 +
    all_rounders["bowling_score"] * 0.5
)

# =========================
# 7. ROLE CLASSIFICATION
# =========================
top_order_batsmen = batting_stats[
    batting_stats["avg_batting_position"] <= 3.5
].sort_values("batting_score", ascending=False).copy()

middle_order_batsmen = batting_stats[
    (batting_stats["avg_batting_position"] > 3.5) &
    (batting_stats["avg_batting_position"] <= 6.5)
].sort_values("batting_score", ascending=False).copy()

# Bowling types by name inference
spinners_list = [
    "chahal", "rashid", "ashwin", "jadeja", "narine", "chakravarthy",
    "bishnoi", "rahul chahar", "theekshana", "sundar", "sai kishore",
    "kuldeep", "noor", "zampa", "mujeeb", "shamsi", "brar", "markande"
]

medium_pacer_list = [
    "bhuvneshwar", "harshal", "shardul", "deepak chahar",
    "mohit", "sandeep", "avesh", "unadkat", "thakur"
]

fast_bowler_list = [
    "bumrah", "shami", "siraj", "rabada", "nortje", "umran",
    "wood", "lockie", "hazlewood", "starc", "patirana", "cummins",
    "arshdeep", "prasidh", "madhwal", "tyagi"
]

def classify_bowler(name):
    n = str(name).lower()
    for x in spinners_list:
        if x in n:
            return "Spinner"
    for x in medium_pacer_list:
        if x in n:
            return "Medium Pacer"
    for x in fast_bowler_list:
        if x in n:
            return "Fast Bowler"
    return "Bowler"

bowling_stats["bowling_type"] = bowling_stats["bowler"].apply(classify_bowler)

spinners = bowling_stats[
    bowling_stats["bowling_type"] == "Spinner"
].sort_values("bowling_score", ascending=False)

medium_pacers = bowling_stats[
    bowling_stats["bowling_type"] == "Medium Pacer"
].sort_values("bowling_score", ascending=False)

fast_bowlers = bowling_stats[
    bowling_stats["bowling_type"] == "Fast Bowler"
].sort_values("bowling_score", ascending=False)

other_bowlers = bowling_stats[
    bowling_stats["bowling_type"] == "Bowler"
].sort_values("bowling_score", ascending=False)

# =========================
# 8. SPECIAL CATEGORY PLAYERS
# =========================
highest_scorer = batting_stats.sort_values("runs", ascending=False).head(1)
highest_wicket_taker = bowling_stats.sort_values("wickets", ascending=False).head(1)

# =========================
# 9. FINAL TEAM SELECTION
# Team = 5 batsmen + 2 all-rounders + 4 bowlers
# =========================
final_team = []
selected_players = set()

def add_player(df, player_col, role_name, score_col, count):
    added = 0
    for _, row in df.iterrows():
        player = row[player_col]
        if player not in selected_players:
            final_team.append({
                "Player": player,
                "Role": role_name,
                "Score": round(row[score_col], 2)
            })
            selected_players.add(player)
            added += 1
            if added == count:
                break

# 5 batsmen
add_player(highest_scorer, "batter", "Highest Scorer Batter", "batting_score", 1)
add_player(top_order_batsmen, "batter", "Top Order Batter", "batting_score", 2)
add_player(middle_order_batsmen, "batter", "Middle Order Batter", "batting_score", 2)

# 2 all-rounders
add_player(all_rounders.sort_values("allrounder_score", ascending=False), "player", "All-Rounder", "allrounder_score", 2)

# 4 bowlers
add_player(highest_wicket_taker, "bowler", "Highest Wicket-Taker", "bowling_score", 1)
add_player(fast_bowlers, "bowler", "Fast Bowler", "bowling_score", 1)
add_player(medium_pacers, "bowler", "Medium Pacer", "bowling_score", 1)
add_player(spinners, "bowler", "Spinner", "bowling_score", 1)

# If any category remains empty, fill with best remaining bowlers/batters
if len(final_team) < 11:
    remaining_batters = batting_stats[
        ~batting_stats["batter"].isin(selected_players)
    ].sort_values("batting_score", ascending=False)

    for _, row in remaining_batters.iterrows():
        if len(final_team) >= 11:
            break
        final_team.append({
            "Player": row["batter"],
            "Role": "Extra Batter",
            "Score": round(row["batting_score"], 2)
        })
        selected_players.add(row["batter"])


if len(final_team) < 11:
    for _, row in all_rounders.sort_values("allrounder_score", ascending=False).iterrows():
        if len(final_team) >= 11:
            break
        if row["player"] not in selected_players:
            add_player(row["player"], "All-Rounder", "Additional All-Rounder")

if len(final_team) < 11:
    for _, row in bowling_stats.sort_values(["bowling_score", "wickets"], ascending=False).iterrows():
        if len(final_team) >= 11:
            break
        if row["bowler"] not in selected_players:
            add_player(row["bowler"], "Bowler", "Additional Best Bowler")

if len(final_team) < 11:
    for _, row in all_rounders.sort_values("allrounder_score", ascending=False).iterrows():
        if len(final_team) >= 11:
            break
        if row["player"] not in selected_players:
            add_player(row["player"], "All-Rounder", "Additional All-Rounder")

if len(final_team) < 11:
    for _, row in bowling_stats.sort_values(["bowling_score", "wickets"], ascending=False).iterrows():
        if len(final_team) >= 11:
            break
        if row["bowler"] not in selected_players:
            add_player(row["bowler"], "Bowler", "Additional Best Bowler")

# =========================
# 10. PRINT FINAL TEAM
# =========================
final_team_df = pd.DataFrame(final_team)

print("\n=========== FINAL TEAM (2020 to 2025) ===========\n")
print(final_team_df.to_string(index=False))
```

```python
# =========================
# REQUIRED DATAFRAMES FOR SELECTION BASIS
# =========================

# Overall best batsmen
overall_best_batsmen = batting_stats.sort_values(
    ["batting_score", "runs"], ascending=False
)

# Top-order batsmen
top_order_batsmen = batting_stats[
    batting_stats["avg_batting_position"] <= 3.5
].sort_values(["batting_score", "runs"], ascending=False)

# Middle-order batsmen
middle_order_batsmen = batting_stats[
    (batting_stats["avg_batting_position"] > 3.5) &
    (batting_stats["avg_batting_position"] <= 6.5)
].sort_values(["batting_score", "runs"], ascending=False)

# Highest wicket-taker
highest_wicket_taker_df = bowling_stats.sort_values(
    ["wickets", "bowling_score"], ascending=False
)

# Bowler type classification
fast_keywords = ["shami", "bumrah", "siraj", "wood", "ferguson", "nortje", "cummins", "starc", "tyagi", "umran", "rabada", "ishant", "harshal"]
medium_keywords = ["bhuvneshwar", "chahar", "thakur", "pandya", "sakariya", "sandeep", "mohit", "avesh", "arshdeep", "deshpande", "unadkat"]
spinner_keywords = ["rashid", "chahal", "ashwin", "jadeja", "axar", "varun", "bishnoi", "kuldeep", "sundar", "hasaranga", "theekshana", "mujeeb", "narine", "rahul chahar"]

def classify_bowler(name):
    n = str(name).lower()
    if any(k in n for k in spinner_keywords):
        return "Spinner"
    elif any(k in n for k in fast_keywords):
        return "Fast Bowler"
    elif any(k in n for k in medium_keywords):
        return "Medium Pacer"
    else:
        return "Bowler"

bowling_stats["bowler_type"] = bowling_stats["bowler"].apply(classify_bowler)

fast_bowlers = bowling_stats[
    bowling_stats["bowler_type"] == "Fast Bowler"
].sort_values(["bowling_score", "wickets"], ascending=False)

medium_pacers = bowling_stats[
    bowling_stats["bowler_type"] == "Medium Pacer"
].sort_values(["bowling_score", "wickets"], ascending=False)

spinners = bowling_stats[
    bowling_stats["bowler_type"] == "Spinner"
].sort_values(["bowling_score", "wickets"], ascending=False)
```

```python
final_team = []
selected_players = set()

def add_player(player_name, role, reason):
    if player_name not in selected_players:
        final_team.append({
            "Player": player_name,
            "Role": role,
            "Selection_Basis": reason
        })
        selected_players.add(player_name)

# 5 BATSMEN
for _, row in overall_best_batsmen.iterrows():
    add_player(row["batter"], "Batsman", "Highest Scorer")
    break

count = 0
for _, row in top_order_batsmen.iterrows():
    if row["batter"] not in selected_players:
        add_player(row["batter"], "Batsman", "Top Order Batsman")
        count += 1
        if count == 2:
            break

count = 0
for _, row in middle_order_batsmen.iterrows():
    if row["batter"] not in selected_players:
        add_player(row["batter"], "Batsman", "Middle Order Batsman")
        count += 1
        if count == 2:
            break

# 2 ALL-ROUNDERS
count = 0
for _, row in all_rounders.sort_values(["allrounder_score", "runs", "wickets"], ascending=False).iterrows():
    if row["player"] not in selected_players:
        add_player(row["player"], "All-Rounder", "Best All-Rounder")
        count += 1
        if count == 2:
            break

# 4 BOWLERS
for _, row in highest_wicket_taker_df.iterrows():
    if row["bowler"] not in selected_players:
        add_player(row["bowler"], "Bowler", "Highest Wicket-Taker")
        break

for _, row in fast_bowlers.iterrows():
    if row["bowler"] not in selected_players:
        add_player(row["bowler"], "Bowler", "Fast Bowler")
        break

for _, row in medium_pacers.iterrows():
    if row["bowler"] not in selected_players:
        add_player(row["bowler"], "Bowler", "Medium Pacer")
        break

for _, row in spinners.iterrows():
    if row["bowler"] not in selected_players:
        add_player(row["bowler"], "Bowler", "Spinner")
        break

# Fill if less than 11
if len(final_team) < 11:
    for _, row in overall_best_batsmen.iterrows():
        if len(final_team) >= 11:
            break
        if row["batter"] not in selected_players:
            add_player(row["batter"], "Batsman", "Additional Best Batter")

if len(final_team) < 11:
    for _, row in all_rounders.sort_values("allrounder_score", ascending=False).iterrows():
        if len(final_team) >= 11:
            break
        if row["player"] not in selected_players:
            add_player(row["player"], "All-Rounder", "Additional All-Rounder")

if len(final_team) < 11:
    for _, row in bowling_stats.sort_values(["bowling_score", "wickets"], ascending=False).iterrows():
        if len(final_team) >= 11:
            break
        if row["bowler"] not in selected_players:
            add_player(row["bowler"], "Bowler", "Additional Best Bowler")

final_team_df = pd.DataFrame(final_team)

print(final_team_df[["Player", "Role", "Selection_Basis"]])
```
